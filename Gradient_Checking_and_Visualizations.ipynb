```python
import matplotlib.pyplot as plt
import numpy as np
import scipy
```

## Linear Regression

```python
data_original = np.loadtxt('stanford_dl_ex/ex1/housing.data')
data = np.insert(data_original, 0, 1, axis=1)
np.random.shuffle(data)
train_X = data[:400, :-1]
train_y = data[:400, -1]
test_X = data[400:, :-1]
test_y = data[400:, -1]

m, n = train_X.shape
theta = np.random.rand(n)

def cost_function(theta, X, y): 
    squared_errors = (X.dot(theta) - y) ** 2
    J = 0.5 * squared_errors.sum()
    return J

def gradient(theta, X, y):
    errors = X.dot(theta) - y
    return errors.dot(X)

def train(X, y):
    return scipy.optimize.minimize(
        fun=cost_function,
        x0=theta,
        args=(X, y),
        method='bfgs',
        jac=gradient,
        options={'maxiter': 200, 'disp': False},
    ).x
```

## Gradient Checking

```python
epsilon = 1e-4
```

```python
mask = np.identity(theta.size)
```

```python
theta_plus = theta + epsilon * mask
theta_minus = theta - epsilon * mask
```

```python
diffs = np.empty_like(theta)
for i in xrange(theta_plus.shape[0]):
    gradient_def = (
        (cost_function(theta_plus[i], train_X, train_y) - cost_function(theta_minus[i], train_X, train_y)) /
        (2.0 * epsilon)
    )
    gradient_lin_reg = gradient(theta, train_X, train_y)[i]
    diffs[i] = np.absolute(gradient_def - gradient_lin_reg)
```

```python
diffs
```

Lookin' good!

```python
assert all(np.less(diffs, 1e-4))
```

## Learning Curve

```python
fractions = 20.0
rms_train = np.zeros(fractions)
rms_test = np.zeros(fractions)
for i, fraction in enumerate(np.linspace(m / fractions, m, fractions)):
    optimal_theta = train(train_X[:fraction], train_y[:fraction])
    rms_train[i] = np.sqrt(np.mean((train_X[:fraction].dot(optimal_theta) - train_y[:fraction]) ** 2))
    rms_test[i] = np.sqrt(np.mean((test_X.dot(optimal_theta) - test_y) ** 2))
```

```python
plt.figure(figsize=(8, 6))
plt.xlim(0, 1.1 * m)
plt.xlabel('Training Instances Used')
plt.ylabel('RMS Error')
plt.plot(np.arange(1 / fractions, 1 + 1 / fractions, 1 / fractions) * m, rms_train, c='g', marker='o', label='Train')
plt.plot(np.arange(1 / fractions, 1 + 1 / fractions, 1 / fractions) * m, rms_test, c='b', marker='o', label='Test')
plt.legend()
```
